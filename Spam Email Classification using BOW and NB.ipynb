```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv('email_spam.csv')
df=pd.DataFrame(d)
```

```python
df.head()
```

```python
df.info()
```

```python
df.Category.value_counts()
```

```python
#creating a numeric column to represent spam and non-spam
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df
```

```python
#independent variable x dependent variable y
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.spam,train_size=0.8,test_size=0.2)
```

```python
df['Message'].shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_cv=v.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
v.vocabulary_
```

```python
#to get the necessary features
dir(v)
```

```python
v.get_feature_names_out()[1771:2000]
```

# Training Using Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
#put training set in the model
model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

Evaluate on the Testing Set

```python
from sklearn.metrics import classification_report
X_test_cv = v.transform(X_test)
y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
# Assuming you have a vectorizer 'v' and a model 'model' for email classification
emails = [
    'Hello Paridhi, can we get together to watch a movie tomorrow tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!',
    'Hey there, how about coffee this afternoon?',
    'Congratulations! You have  won a free iPhone. Click here to claim your prize!',
]

# Transform the emails into the same format as your training data
emails_count = v.transform(emails)

# Predict the class labels for the emails
predicted_labels = model.predict(emails_count)

# Loop through the predicted labels and print "Ham" or "Spam" accordingly
for label in predicted_labels:
    if label == 0:
        print('Ham')
    else:
        print('Spam')
```
